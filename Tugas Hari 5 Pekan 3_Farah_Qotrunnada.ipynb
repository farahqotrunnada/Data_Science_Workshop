````python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
````

````python
df = pd.read_csv('https://raw.githubusercontent.com/vaksakalli/datasets/master/diamonds.csv')
df.head()
````

# Soal 1: Scatter Plot

Buatlah scatterplot menggunakan Seaborn dari data df di atas, variasikan menggunakan palette yang telah disediakan sehingga didapatkan output sesuai expected.

````python
fig, ax = plt.subplots(figsize=(13, 13))
palette="ch:r=-.2,d=.3_r"

#code 
sns.set(style='white')
ax = sns.scatterplot(df["carat"], df["price"], palette=palette, hue=df['clarity'], edgecolor='None', size=df['depth'],sizes=(1, 10))

ax.set_xlabel('carat')
ax.set_ylabel('price')
plt.legend(loc='upper left')
ax.set_frame_on(False) 

plt.show()
````

````python
#code here
def hexbin(x, y, color, **kwargs):
    cmap = sns.light_palette(color, as_cmap=True)
    plt.hexbin(x, y, gridsize=15, cmap=cmap, **kwargs)
    
grid=sns.FacetGrid(df, hue='cut',col='cut',height=4)
grid=grid.map(hexbin, 'carat','price', extent=[0,3,0,20000])
````

# Soal 3: Pair Plot

Ambil 300 data secara acak dari df, kemudian buat pair plotnya dari 300 data tersebut untuk kolom carat, depth, table, dan price sehingga dihasilkan output sesuai expected.

````python
df = df.sample(n=300, random_state=123)
plt.style.use('default')

#code here
sns.pairplot(df,vars=['carat','depth','table','price'],hue='cut',hue_order=['Premium','Ideal','Very Good','Good','Fair'], corner=True, markers=['+','o','s','D','x'])
````

# Soal 4: Joint Plot

Gunakan data yang sama dengan soal 3 untuk menghasilkan dua visualisasi joint plot antara carat dan price nya sesuai expected output.

````python
#code here
import scipy.stats
with sns.axes_style('white'):
  sns.jointplot('carat','price',data=df,kind='kde')
  sns.jointplot('carat','price',data=df,kind='reg',stat_func=scipy.stats.pearsonr)
````
